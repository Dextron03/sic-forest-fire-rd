```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_forest_fires = pd.read_csv("../data/Incendios-Forestales-2021.csv")
```

## ¿Cuáles son las causas más comunes de los incendios forestales?

```python
df_causes = df_forest_fires.loc[:, ["Causas"]]
```

```python
df_causes
```

```python
causes = []

values_df_causes : list = df_causes.values

for value_causes in values_df_causes:
    causes.append("".join(value_causes))
```

**Causas en una unidad:**

```python
causas = []

for item in causes:
    if causas.count(item) >= 1:
        pass
    else:
        causas.append(item)
        
```

**Cantidad de causas:**

```python
total_of_causes = []
for item in causas:
    total_of_causes.append(causes.count(item))
```

**Grafica:**

```python
plt.barh(causas, total_of_causes)

# Agregar etiquetas y título
plt.xlabel('Cantidades de veces')
plt.ylabel('Causas')
plt.title('Frecuencia de Ocurrencia por Causa')

# Mostrar el gráfico
plt.show()
```

## ¿Cuáles son los municipios/provincias más afectados por los incendios forestales?

### MUNICIPIOS

```python
df_forest_fires
```

```python
df_forest_fires = df_forest_fires.loc[:, ["Municipio", "Provincia"]]
```

```python
df_forest_fires
```

```python
df_all_municipalities =  df_forest_fires["Municipio"]
```

```python
df_municipalities =  df_forest_fires["Municipio"].drop_duplicates()
```

```python
numbers_total = []
for item in df_municipalities:
    numbers_total.append(list(df_all_municipalities).count(item))
```

```python
plt.barh(list(df_municipalities), numbers_total)
plt.subplots_adjust(left=0.1, right=0.9, top=1.5, bottom=0.1)

# Agregar etiquetas y título
plt.xlabel('Cantidades de incendios')
plt.ylabel('Municios')
plt.title('Cantidad de Incendios segun el Municipio.')

# Mostrar el gráfico
plt.show()
```
